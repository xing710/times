```python
import pandas as pd
import numpy as np
import jieba
import jieba.posseg as pseg
from wordcloud import WordCloud
import matplotlib.pyplot as plt


filename = "HLM.txt" #文件路径
stopwords_file = "stop_words.txt"
# with open(filename,encoding="utf-8") as f:
#  mytext = f.read()
text= open(filename,encoding="utf-8").read()  #打开文件
font = r'C:\Users\Dell\AppData\Local\Microsoft\Windows\Fonts\stfangso.ttf' #字体库路径

stop_f = open(stopwords_file,"r",encoding='utf-8') #打开stop words
stop_words = list()

for line in stop_f.readlines():
    line = line.strip()
    if not len(line):
        continue
    stop_words.append(line)  #将stop words 读入内存中
stop_f.close
# print(len(stop_words))
result = list()
outstr = '' 
# mytext = " ".join(jieba.cut(text,cut_all=False))
mytext = pseg.cut(text)
# print(mytext)
for word,flag in mytext:          #filter stop words
#         print (word)
        if word not in stop_words:  
            outstr += word 
            outstr += " "
#         else:
# #             print(word)
result.append(outstr.strip())

for word in result:
#     print(word)
    outstr +=word
print (type(outstr))    
# print(result)
wordcloud = WordCloud(font_path=font).generate(outstr)  #生成词云图

data =wordcloud.process_text(outstr)  # 打印wordcloud中的关键字和出现频次,去掉停止词
s = pd.DataFrame([['test',1]],columns =['names','times'])  #建立一个 pd数据
# i = lens(s)
# print(data.items)
i = 0
for x ,y in data.items():  
     s.loc[i] =[x,y]      #从字典中提取数据，存储到pd中
     i +=1
# print(s.sort_values(by='times',ascending=False)[:10])
t =s.sort_values(by='times',ascending=False)[:20]
t.index = range(len(t))  #重新拍列序号
print(t)
```
